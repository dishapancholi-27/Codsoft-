```python
encoder = LabelEncoder()
Movie_genre_data_combined["category"] = encoder.fit_transform(Movie_genre_data_combined["category"].values)
```

```python
encoder = LabelEncoder()
Movie_genre_data_combined["movie_name"] = encoder.fit_transform(Movie_genre_data_combined["movie_name"].values)
```

```python
Movie_genre_data_combined.category = Movie_genre_data_combined.category.fillna(Movie_genre_data_combined.category.mean())
```

# Preprocess the Data

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(Movie_genre_data_combined["confession"])
```

```python
Movie_genre_data_combined.head()
```

```python
Y = Movie_genre_data_combined["category"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

# Train Naive Bayes Classifier

```python
naive_bayes_model = MultinomialNB()
```

```python
naive_bayes_model.fit(X_train, Y_train)
```

```python
nb_predictions = naive_bayes_model.predict(X_test)
```

```python
print("Naive Bayes Model:")
print(confusion_matrix(Y_test, nb_predictions))
print(classification_report(Y_test, nb_predictions))
print("Accuracy:", accuracy_score(Y_test, nb_predictions))
print("r2_score:", r2_score(Y_test, nb_predictions))
```

# Train Logistic Regression Model

```python
logistic_regression_model = LogisticRegression()
```

```python
logistic_regression_model.fit(X_train, Y_train)
```

```python
lr_predictions = logistic_regression_model.predict(X_test)
```

```python
print("Logistic Regression Model:")
print(confusion_matrix(Y_test, lr_predictions))
print(classification_report(Y_test, lr_predictions))
print("Accuracy:", accuracy_score(Y_test, lr_predictions))
print("r2_score:", r2_score(Y_test, lr_predictions))
```
